### Defining plot color

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,1,0.5]

plt.bar(x,y)
```

```python
wide = [0.5,0.5,0.5,0.7,1,0.7,0.5,0.5,0.5,]
color = ['salmon']

plt.bar(x, y, width=wide, color=color, align='center')
```

### Customizing line styles

```python
x1 = range(1,10)
y1 = [9,8,7,6,5,4,3,2,1]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
x2 = range(1,10)
y2 = [2,4,6,8,6,4,2,4,6] 

plt.plot(x, y, ls='solid', lw=3)
plt.plot(x1, y1, ls='--', lw=7)
plt.plot(x2, y2, ls='-.')
```

### Setting plot markers

```python
plt.plot(x, y, marker='1', mew=15)
plt.plot(x1, y1, marker='+', mew=10)
plt.plot(x2, y2, marker='d', mew=1)
```
